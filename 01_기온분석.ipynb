```python
import matplotlib.pyplot as plt

plt.title("The Highest Temperature in Seoul")
plt.plot(range(len(date)),temp,linewidth=0.5,color='green')
plt.show()
```

## 02. 7월과 8월 중 서울최고기온 분석

```python
import csv
f=open('seoul_temp.csv','r')
data=csv.reader(f)

next(data)
#a list of the highest temperatures per day in July and August throughout history
temp=[]
date=[]
#x=[]
year=[]
for i in data:
    if i[-1]!='' and (int(i[0].split('-')[1])==7 or int(i[0].split('-')[1])==8):
        #if i[0].split('-')[1]=='07' or i[0].split('-')[1]=='08':
        temp.append(float(i[-1]))
        date.append(i[0])
        year.append(int(i[0].split('-')[0]))
#for i in date:
#    x.append(i.split('-')[0])
#year=[int(j) for j in x]
print(max(temp))
pos=temp.index(max(temp))
print(date[pos])

day=date[pos].split('-')
print("서울최고기온은 {}이고, {}년 {}월 {}일입니다.".format(max(temp),day[0],day[1],day[2]))
```

```python
import matplotlib.pyplot as plt

plt.title("The Highest Temperatures in Seoul in July and August")
plt.plot(year,temp,linewidth=0.5,color='red')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.show()
```

## 03. 자신의 생일을 입력하고 생일 날 가장 더웠던 날과 추웠던 날을 표시하시오.

```python
import csv
f=open('seoul_temp.csv','r')
month=int(input('몇 월 생인가요?: '))
day=int(input('며칠인가요?: '))

data=csv.reader(f)
next(data)
#a list of the highest temperatures on my birthdays every year
max_temp=[]
#a list of the lowest temperatures on my birthdays every year
min_temp=[]
year_hot=[]
year_cold=[]
for i in data:
    if i[-1]!='' and (int(i[0].split('-')[1])==month and int(i[0].split('-')[2])==day):
        max_temp.append(float(i[-1]))
        year_hot.append(int(i[0].split('-')[0]))
    if i[-2]!='' and (int(i[0].split('-')[1])==month and int(i[0].split('-')[2])==day):
        min_temp.append(float(i[-2]))
        year_cold.append(int(i[0].split('-')[0]))
print('가장 더웠던 날:',max(max_temp))
print('가장 추웠던 날:',min(min_temp))
pos=max_temp.index(max(max_temp))
pos2=min_temp.index(min(min_temp))
print("가장 더웠던 해는 {}년이고, 가장 추웠던 해는 {}년이다.".format(year_hot[pos],year_cold[pos2]))
f.close()
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi=100)
plt.title('Temperatures on My Birthdays')
plt.plot(year_hot,max_temp,color='r',label='highest temp',linewidth=0.5)
plt.plot(year_cold,min_temp,color='b',linewidth=0.5,label='lowest temp')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.legend(loc=4)
plt.show()
f.close()
```
